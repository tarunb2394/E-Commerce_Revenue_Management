## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

```python
ECom_Data["final_price"] = ECom_Data["discounted_price"].fillna(ECom_Data["retail_price"])

# Define a function to compute the revenue for each order
def compute_revenue(row):
    if row["final_price"] > 600:
        return row["final_price"] * 0.25
    elif row["final_price"] > 350:
        return row["final_price"] * 0.15
    elif row["final_price"] > 100:
        return row["final_price"] * 0.1
    else:
        return row["final_price"] * 0.05
    
    # Compute the revenue for each order
ECom_Data["revenue"] = ECom_Data.apply(compute_revenue, axis=1)

# Sum the revenue values to get the total revenue generated
total_revenue = ECom_Data["revenue"].sum()

print("Total revenue generated by the company:", total_revenue)
```

## Q6.Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order (6 marks)
#### Hint: Total BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's revenue.
#### Brand Revenue is the discounted price minus the Revenue (generated using the conditions given above Q5)


### Ans 6

```python
#Revenue for each retailer (Brands)
ECom_Data["final_price"] = ECom_Data["discounted_price"].fillna(ECom_Data["retail_price"])

# Define a function to compute the revenue for each order
def compute_revenue(row):
    if row["final_price"] > 600:
        return row["final_price"] * 0.25
    elif row["final_price"] > 350:
        return row["final_price"] * 0.15
    elif row["final_price"] > 100:
        return row["final_price"] * 0.1
    else:
        return row["final_price"] * 0.05

# Group the dataset by Brand and compute the revenue for each order
grouped = ECom_Data.groupby("Brand")
revenue_by_brand = grouped.apply(lambda x: x.apply(compute_revenue, axis=1).sum())

print("Revenue earned by each brand:")
print(revenue_by_brand)
```

```python
ECom_Data["revenue_generated"] = ECom_Data.apply(compute_revenue, axis=1)
ECom_Data["BrandRevenue"] = ECom_Data["discounted_price"] - ECom_Data["revenue_generated"]

# Group the dataset by Brand and calculate the total BrandRevenue for each brand
grouped = ECom_Data.groupby("Brand")
Total_BrandRevenue = grouped["BrandRevenue"].sum()

print('The total brand revenue is:',Total_BrandRevenue.sum())

# Sort the BrandRevenue in descending order and print the top 10 brands with the maximum revenue
top_10_brands = ECom_Data["BrandRevenue"].sort_values(ascending=False).head(10)
print("Top 10 brands with maximum revenue:")
print(top_10_brands)
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

```python
ECom_Data["Order_Date"] = pd.to_datetime(ECom_Data["Order_Date"], infer_datetime_format=True)
revenue = ECom_Data.groupby(["Brand", "Region", pd.Grouper(key="Order_Date", freq="M")])["BrandRevenue"].sum().reset_index()

# extract the month from the Order_Date column
revenue["Month"] = revenue["Order_Date"].dt.to_period("M")

# create a pivot table with Brand as rows, Month as columns, and Region as data
pivot = revenue.pivot_table(values="BrandRevenue", index=["Brand", "Month"], columns=["Region"], fill_value=0)

# Sort the dataframe by Month
pivot = pivot.sort_values("Month")

# In "MM-YYYY" Format
pivot["Month"] = pivot.index.get_level_values("Month").strftime("%b-%y")

# Plot the pivot table as line plots
pivot.set_index("Month").plot(kind="line", figsize=(12,8))
plt.title("Monthly revenue by Brand and Region")
plt.ylabel("Revenue")
plt.show()
```

```python
revenue.head()
```

### Ans 9 b)

```python
# Group the data by region and month, and calculate total revenue
region_month_revenue = ECom_Data.groupby(['Region', pd.Grouper(key='Order_Date', freq='M')])['BrandRevenue'].sum().reset_index()

# Find the best performing month for each region
best_month = region_month_revenue.sort_values(['Region', 'BrandRevenue'], ascending=[True, False]).groupby('Region').first().reset_index()[['Region', 'Order_Date']]

worst_month = region_month_revenue.sort_values(['Region', 'BrandRevenue'], ascending=[True, True]).groupby('Region').first().reset_index()[['Region', 'Order_Date']]


print("Best performing month for each region:")
print(best_month)
print("\nWorst performing month for each region:")
print(worst_month)
```
